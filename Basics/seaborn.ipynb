```python
# joinplot, pairplot , distplot  
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()       
```

```python
# Correlation -- Only For floating and int dtypes
# df.dtypes
df.corr()    
```

```python
sns.heatmap(df.corr())
```

```python
# Joint plot  --
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

```python
sns.pairplot(data = df, hue = 'smoker')
```

```python
#  Find how many males and females are smokers?
df[(df.sex == 'Male') & (df.smoker == 'Yes')].shape[0]  # Male Smokers 
df[(df.sex == 'Female') & (df.smoker == 'Yes')].shape[0] # Female Smokers
```

```python
df.smoker.value_counts()
```

```python
#  Distplot
sns.distplot(df.tip)      # Y- axis : Density 
```

```python
sns.distplot(df.tip, kde = False, bins = 10)   # Y - axis: Count
```

```python
#  Boxplot, Violinplot, Countplot, Barplot 
df.head()
```

```python
df.dtypes
```

```python
sns.countplot('sex', data = df)
```

```python
sns.countplot('smoker', data = df)
```

```python
sns.countplot('day', data = df)
```

```python
sns.countplot(y = 'day', data = df)
```

```python
# Barplot :  need to specify x and y parameters
sns.barplot(x = 'total_bill', y = 'sex', data = df)
```

```python
sns.barplot(x = 'smoker', y = 'total_bill', data = df)
```

```python
# Boxplot  -- helps us to determine the data distribution and outliers
sns.boxplot('sex', 'total_bill', data = df)
```

```python
sns.boxplot('smoker', 'total_bill', data = df, palette = 'rainbow')  # for colors use palette
```

```python
sns.boxplot('total_bill', 'day', hue = 'smoker', data = df)  # 
```

```python
#  Violinplot  -- distribution of data in terms of Kernel Density Estimate and Boxplot
sns.violinplot('total_bill', 'day', data = df, palette = 'rainbow')
```
