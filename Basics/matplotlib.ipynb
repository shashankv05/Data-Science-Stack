```python
x = np.arange(0, 2 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.title('sine wave')
```

```python
x = np.arange(0, 2 * np.pi, 0.1)
y = np.sin(x)
plt.subplot(1, 1, 1)
plt.plot(x, y, 'r')
y = np.cos(x)
# plt.subplot(1, 1, 1)
plt.plot(x, y, 'g') 
plt.title('Sine Vs Cosine')
```

```python
# Bar Plot
x1 = [1, 5, 9]
y1 = [5, 25, 45]               # y = 5 * x
plt.bar(x1, y1, color = 'g')

x2 = [2, 7, 11]
y2 = [2, 7, 11]              # y = x
plt.bar(x2, y2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar plot')
```

```python
# Histogram
x = np.array([50, 95, 40, 45, 75, 100,40])            # Range 40 - 100
plt.hist(x, bins = 10)                                  
plt.title('Histogram Plot')
```

```python
x = np.array([10, 59, 61, 40, 62, 69, 80, 95, 97])    # Range 10 - 100
plt.hist(x, bins = 10)
plt.title('Histogram ')
```

```python
#  Boxplot 
# data = np.random.normal(0, 1, 100)
# data
# data = np.random.normal(0, 2, 100)
# data

data =[np.random.normal(0, sd, 100) for sd in range(1, 4)]
data
plt.boxplot(data, vert = True, patch_artist = True)
plt.title('Boxplot')    # Circles in the plot refers to Outliers
```

```python
#  PieChart 
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [259, 201, 170, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True)
plt.title('Pie Chart')
```
