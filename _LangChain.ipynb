```python
from langchain import PromptTemplate, HuggingFaceHub, LLMChain
```

**In language models, the temperature parameter controls the randomness or creativity of the generated text.                 
Higher temperature (e.g., above 1.0) leads to more diverse and random outputs, while a lower temperature (e.g., below 1.0) produces more focused and deterministic outputs.**

**Using Hugging-Face**

```python
flan_t5 = HuggingFaceHub(
    repo_id = "google/flan-t5-xl", 
    model_kwargs = {"temperature" : 1e-10}
)
```

```python
flan_t5
```

```python
# prompt template
template = """Question: {question}
Answer: """
prompt = PromptTemplate(template = template, input_variables=["question"])
```

```python
llm_chain = LLMChain(
    prompt = prompt,
    llm = flan_t5
)
```

```python
llm_chain
```

```python
question = "Which NFL team won the super bowl in 2010?"
```

```python
llm_chain.run(question)
```

```python
qs = [
        {"question": "Who directed the movie 'The Shawshank Redemption'?"},
        {"question": "Which actor played the iconic character of Tony Stark, also known as Iron Man, in the Marvel Cinematic Universe?"},
        {"question": "Who directed the movie 'The Dark Knight'?"},
        {"question": "Who portrayed the character of Wolverine in the 'X-Men' film series?"}
    ]
```

```python
llm_chain.generate(qs)
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-xl")
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-xl", device_map="auto")
```

```python
input_text = "translate English to German: How old are you?"
input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to("cuda")

outputs = model.generate(input_ids)
tokenizer.decode(outputs[0])
```

```python
_outtext = tokenizer.decode(outputs[0])
```

```python
import re
```

```python
re.sub(r'<[^>]+>', '', _outtext).strip()
```
